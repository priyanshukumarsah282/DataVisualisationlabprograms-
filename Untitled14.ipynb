```python
import numpy as np
import matplotlib.pyplot as plt

# clean review_count
df['review_count'] = df['review_count'].astype(str).str.replace(',', '')
df['review_count'] = pd.to_numeric(df['review_count'], errors='coerce')

# take top 50 again after cleaning
top50 = df.sort_values(by='employees', ascending=False).head(50)

# remove invalid values
top50 = top50.dropna(subset=['review_count'])
top50 = top50[top50['review_count'] > 0]
```

```python
plt.figure(figsize=(8,6))

plt.hist(top50['review_count'], bins='auto')

plt.title('Histogram of Review Count (Top 50 Companies)')
plt.xlabel('Review Count')
plt.ylabel('Frequency')

plt.grid(axis='y')
plt.show()
```
